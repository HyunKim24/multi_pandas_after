```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# 글꼴을 바꿔라
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
# 유니코드를 적용해라
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# sorted(x) => 문자열로 구성된 리스트를 정렬해준다.
x = ['홍길동','이순신','임꺽정','이이']
print(sorted(x))
y = [90,77,50,80]
plt.bar(sorted(x),y)
plt.show()
```

```python
# 시리즈면 편리함
x =pd.Series(['홍길동','이순신','임꺽정','이이']) 
y = [90,77,50,80]
plt.bar(x.sort_values(),y)
plt.show()
```

```python
# 뒤쪽에 가려져있음
xLabel = ['홍길동','이순신','임꺽정','이이']
x = pd.Series([1,2,3,4])
kor = [90,77,50,80]
eng = [60,100,50,20]
# plt.xticks(x+0.2,xLabel) => x축 값은 xLabel로 대체하라(x+a 는 그래프 조정)
plt.xticks(x+0.2,xLabel)
plt.bar(x, kor, color='r',width=0.4,label = '국어')
plt.bar(x+0.4, eng, color='b',width=0.4,label = '영어')
plt.xlabel("이름")
plt.ylabel("점수")
plt.title('1학년 성적')
plt.legend() # plt.legend(['국어','영어'])
plt.show()
```

```python
xLabel = ['홍길동','이순신','임꺽정','이이']
x = [1,2,3,4]
kor = [90,77,50,80]
eng = [60,100,50,20]

total = list()
for i,v in zip(kor, eng):
    total.append(i+v)
    
plt.xticks(x,xLabel)

plt.bar(x, kor, color='r',width=0.4,label = '국어',alpha = 0.5)
plt.bar(x, eng, color='b',bottom = kor,width=0.4,label = '영어',alpha = 0.5)

for x, kor,eng, total in zip(x, kor,eng,total):
    plt.text(x,kor/2,((kor)/total)*100,va='center',ha='center')
    plt.text(x,kor+eng/2,((eng)/total)*100,va='center',ha='center')
    
plt.xlabel("이름")
plt.ylabel("점수")
plt.title('1학년 성적')
plt.legend() # plt.legend(['국어','영어'])
plt.show()
```

```python
xLabel = ['홍길동','이순신','임꺽정','이이']
x = pd.Series( [1,2,3,4] )
kor = [90,77,50,80]
eng = [60,100,50,20]

for i in range( len(xLabel) ):
    total = kor[i]+eng[i]
    kor_r = kor[i]/total*100
    eng_r = eng[i]/total*100
    plt.text(i+1,kor[i]/2,"%d %%"%kor_r,va='center',ha='center')
    plt.text(i+1,(eng[i]/2)+kor[i],"%d %%"%eng_r,va='center',ha='center')

plt.xticks( x+0.2, xLabel )
plt.bar( x,kor,color='r',width=0.4,label='국어',alpha = 0.5 )
plt.bar( x,eng,color='b',width=0.4,label='영어',bottom=kor,alpha = 0.5 )
plt.xlabel('이름')
plt.ylabel('점수')
plt.title('1학년 성적')
plt.legend()
plt.show()
```
