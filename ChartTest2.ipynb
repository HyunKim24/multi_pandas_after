### 비만도 공식
* 표준체중(kg) = {신장(cm)-100} x 0.85
* 비만도(%) = 현재체중 / 표준체중 x100

* 비만도가
>90이하 : 저체중,    
>90~110: 정상,    
>110~120 : 과체중,    
>120이상 : 비만   

```python
sList = []
colors = []
def obesityFn(h,w):
    stdW = (h-100)*0.85
    biman = (w/stdW)*100
    if biman<=90:
        sList.append(100)
        colors.append('y')
        return '저체중'
    elif biman>=90 and biman<=110:
        sList.append(200)
        colors.append('b')
        return '정상'
    elif 110<biman<=120:
        sList.append(300)
        colors.append('m')
        return '과체중'
    else:
        sList.append(500)
        colors.append('r')
        return '비만'
        
```

```python
obesityFn(170,55)
```

```python
# 원본 데이터
height = [165,177,160,180,185,155,172]
weight = [62,67,55,74,90,43,64]

# 텍스트 리스트 생성
resultList = list()
for h,w in zip(height,weight):
    resultList.append(obesityFn(h,w))

# 각 체중별 차트 사이즈 리스트 형성
sList = list()
for r in resultList:
    if r == '저체중':
        sList.append(100)
    elif r == '정상':
        sList.append(500)
    elif r == '과체중':
        sList.append(1000)
    else:
        sList.append(3000)

# 동일        
colors=['r','g','b','k','m','y','c']
plt.scatter(height,weight,s = sList,c=colors, alpha=0.5)
plt.xlabel('키')
plt.ylabel('몸무게')
plt.grid(True)

# 생성된 리스트 돌려가며 텍스트 찍어
for x,y,r in zip(height,weight,resultList):
    plt.text(x,y,'%s' %r)

# 차트보여줘
plt.show()




#     print(x,y,r)
```
