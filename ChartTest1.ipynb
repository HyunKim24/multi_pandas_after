```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# 글꼴을 바꿔라
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
# 유니코드를 적용해라
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터는 y축으로 인식 
# x축은 인덱스라고 생각하면 쉬움
data1 = [10,14,19,20,25]
plt.plot(data1)
plt.show()
```

```python
# x축과 y축을 모두 설정해 줄수 있음 // 각각을 리스트로 만들어서 plot함수에다가
# 쳐 넣으면 됨
y = [10,14,19,20,25]
x = [1,2,3,4,5]
plt.plot(x,y)
plt.show()
```

```python
# 한개의 차트에 다양한 데이터를 집어넣을때 각각의 데이터를
# 리스트로 넣고 그냥 다 plot함수에다가 넣으면 된다.
x = [1,2,3,4,5]
y1= [10,14,19,20,25]
y2= [5,8,10,25,20]
y3= [30,5,20,11,15]
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

```python
# figure() 하나의 윈도우를 나타냄(없으면 디폴트로 생성되서 하나에 차트에 넣어줌)
x = [1,2,3,4,5]
y1= [10,14,19,20,25]
y2= [5,8,10,25,20]
# plt.figure() # 하나의 윈도우를 나타냄
plt.plot(x,y1)
plt.figure(figsize=[3,3]) # 하나의 윈도우를 나타냄
# figsize=[3,3] = 창의 크기를 설정할 수 있음(인치단위로 나타냄)(default 6*5)
plt.plot(x,y2)
plt.show()
```

```python
#  plt.subplot(2,2,3) => 하나의 창을 2행 2열로 분할하고,3번째에다가 그래프를
# 넣어라(맨 마지막 숫자는 좌에서 우로 간다.) => 하나의 창을 나눈다.
x = [1,2,3,4,5]
y1= [10,14,19,20,25]
y2= [5,8,10,25,20]
y3= [30,5,20,11,15]
# plt.figure()
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.show()
```

```python
x = [1,2,3,4,5]
y1= [10,14,19,20,25]
y2= [5,8,10,25,20]
y3= [30,5,20,11,15]
# plt.figure()
plt.subplot(1,3,1)
plt.plot(x,y1)
plt.subplot(1,3,2)
plt.plot(x,y2)
plt.subplot(1,3,3)
plt.plot(x,y3)
plt.show()
```

```python
# plt.ylim(0,30) => y축의 범위를 설정해 주는 함수(디폴트는 최소값 최대값임)
# plt.xlim(0,10) => x축의 범위를 설정해 주는 함수(디폴트는 최소값 최대값임)
x = [1,2,3,4,5]
y1= [10,14,19,20,25]
plt.plot(x,y1)
plt.ylim(0,30)
plt.xlim(0,10)
plt.show()
```

```python
# plt.plot(x,y1,'ro--') =>'r'은 fmt이므로 선의 색깔을 지정 / o는 마커 / -는 솔리드라인
# plt.xlabel('x축라벨') => x축 라벨의 이름
# plt.ylabel('y축라벨') => y축 라벨의 이름
# plt.title('타이틀') => chart title
# plt.grid(True) => 차트내 격자로 보조선 그림
# plt.legend(['국어','영어','수학']) => plt.plot 이후에 설정하면 각 그래프가 나타내는
# 것들을 요약한다.
x = [1,2,3,4,5]
y1= [10,14,19,20,25]
y2= [5,8,10,25,20]
y3= [30,5,20,11,15]
plt.xlabel('x축라벨')
plt.ylabel('y축라벨')
plt.title('타이틀')
plt.grid(True)
plt.plot(x,y1,'ro--',x,y2,'b>-',x,y3,'g1-')
plt.legend(['국어','영어','수학'])
plt.show()
```

```python
# plt.text(1,10,'10점') => x축의 1 y축의 10 위치에 뒤에 있는 string을 붙여라
# 왠지 for가 편할듯
x = [1,2,3,4,5]
y1= [10,14,19,20,25]
plt.xlabel('x축라벨')
plt.ylabel('y축라벨')
plt.title('타이틀')
plt.grid(True)
plt.plot(x,y1,'ro--')
plt.legend(['국어'])
for x,y in zip(x,y1):
    plt.text(x,y,'%d점' %y)
plt.show()
```
