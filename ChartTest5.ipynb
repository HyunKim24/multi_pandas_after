```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# 글꼴을 바꿔라
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
# 유니코드를 적용해라
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
fruit = ['사과','바나나','딸기','오렌지','포도']
cnt = [7,6,3,2,2]
exp = [0.2, 0.1, 0, 0, 0]

# autopct='%.1f%%' => 퍼센트를 붙여봅시다
# explode=exp => 간격만큼 따로 떨어지게 끔 해 봅시닼ㅋㅋㅋㅋ
# shadow=True => 입체감 ㄱㄱ
plt.pie(cnt,labels=fruit,autopct='%.1f%%',explode=exp,shadow=True)

plt.show()
```
