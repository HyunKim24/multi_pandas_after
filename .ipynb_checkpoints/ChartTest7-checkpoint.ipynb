```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# 글꼴을 바꿔라
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
# 유니코드를 적용해라
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
data = {'kor':[80,30,50,70],'eng':[20,90,90,80],'math':[33,44,55,66]}
df = pd.DataFrame(data,index=['aa','bb','cc','dd'])
df
```

### 연습문제
* 1. 국어, 영어 점수에 대한 바차트를 그리시오
* 2. 영어, 수학 점수에 대한 boxplot을 그리시오 
* 3. 이름별 국어점수를 파이차트로 그리시오
* 4. 국어, 영어, 수학 점수를 라인차트로 그리시오
* 5. 수학점수에 대한 scatter를 그리고, 점수를 text로 표시하시오
* 6. 국어, 영어, 수학점수를 stacked bar로 그리시오(% 표시할것)
* 7. births.txt 데이터의 2000년대 남아 여아 출생 현황을 바차트와 라인차트를 같이 표시하시오(plt.xtics(rotation="45")를 적용하여 x축 데이터를 읽기 좋게 할것)
* 8. hotel.xlsx 데이터를 이용하여 호텔별 가격현황을 바차트로 그리시오
* 9. 국어를 x축으로, 영어를 y축으로 하여 scatter를 그리고 각 원에 이름을 표시하시오


```python
# * 1. 국어, 영어 점수에 대한 바차트를 그리시오
xLabel = df.index
x = pd.Series([1,2,3,4])
kor = df['kor']
eng = df['eng']
plt.xticks(x+0.2,xLabel)
plt.bar(x, kor, color='r',width=0.4,label = '국어')
plt.bar(x+0.4, eng, color='b',width=0.4,label = '영어')
plt.xlabel("이름")
plt.ylabel("점수")
plt.legend() # plt.legend(['국어','영어'])
plt.show()
```

```python
# * 2. 영어, 수학 점수에 대한 boxplot을 그리시오 
eng = df['eng']
math = df['math']
plt.boxplot([eng,math],labels=['영어','수학'])
plt.show()
# 이상치(outlier) => 모여있는 데이터보다 월등히 데이터가 다를때 표시함(math 동그라미)
```

```python
# * 3. 이름별 국어점수를 파이차트로 그리시오
korList = df['kor']
nameList = df.index
plt.pie(korList,labels=nameList,autopct='%.1f%%')
plt.title('국어점수')
plt.show()
```

```python
# * 4. 국어, 영어, 수학 점수를 라인차트로 그리시오
x = df.index
y1= df['kor']
y2= df['eng']
y3= df['math']
plt.xlabel('이름')
plt.ylabel('점수')
plt.title('성적')
plt.grid(True)
plt.plot(x,y1,'ro--',x,y2,'b>-',x,y3,'g1-')
plt.legend(['국어','영어','수학'])
plt.show()
```

```python
# * 5. 수학점수에 대한 scatter를 그리고, 점수를 text로 표시하시오
mathList = df['math']
nameList = df.index
plt.scatter(df.index,mathList,s=500,alpha=0.5) # 산점도 s는 사이즈임 리스트 줄수있어
plt.xlabel('이름')
plt.ylabel('점수')
plt.grid(True)
for h,w in zip(nameList,mathList):
    plt.text(h,w,'%d점'%w)
# plt.show()
```

```python
# * 6. 국어, 영어, 수학점수를 stacked bar로 그리시오(% 표시할것)
student = pd.Series(df.index)
score = df.columns
kor=df['kor']
eng = df['eng']
math=df['math']
plt.bar(student, kor, color='r', width=0.4, label='korean', alpha=0.5)
plt.bar(student, eng, color='g', bottom=kor, width=0.4, label='english', alpha=0.5)
plt.bar(student, math, color='b', bottom=kor+eng, width=0.4, label='math', alpha=0.5)
for i in  range(0, len(student)):
    plt.text(i, kor[i]/2, '%d%%' %(kor[i]/(kor[i]+eng[i]+math[i])*100), va='center', ha='center')
    plt.text(i, kor[i]+eng[i]/2, '%d%%' %(eng[i]/(kor[i]+eng[i]+math[i])*100), va='center', ha='center')
    plt.text(i, +kor[i]+eng[i]+math[i]/2, '%d%%' %(math[i]/(kor[i]+eng[i]+math[i])*100), va='center', ha='center')
    
    


plt.xlabel('name')
plt.ylabel('score')
plt.title('score')
plt.legend()
plt.show()
```

```python
# 7. births.txt 데이터의 2000년대 남아 여아 출생 현황을 바차트와 라인차트를 같이 표시하시오(plt.xtics(rotation="45")를 적용하여 x축 데이터를 읽기 좋게 할것)
dfBirthB=pd.read_table('births.txt', sep=',', header=None, index_col=0 )
dfBirth = dfBirthB.loc['2000':]
dfBirth.columns=['남아수', '여아수']
dfBirth.index.name='년도'
dfBirth

birthYear = pd.Series(dfBirth.index)
girlBirth = dfBirth['여아수']
boyBirth = dfBirth['남아수']
plt.xticks(birthYear, rotation='45')
plt.plot(birthYear,girlBirth, 'ro--', birthYear,boyBirth,'b>-')
plt.bar(birthYear, girlBirth, color='g', width=0.4, label='girls', alpha=0.5)
plt.bar(birthYear, boyBirth, color='red', width=0.4, label='boys', alpha=0.5)

plt.show()
```

```python
# * 8. hotel.xlsx 데이터를 이용하여 호텔별 가격현황을 바차트로 그리시오
hotel = pd.read_excel('hotel.xlsx')
df = pd.DataFrame(hotel)
x =df.index
y = df['price']
plt.xticks(rotation='45')
plt.bar(x,y,width=0.5)
plt.show()
```

```python
# * 9. 국어를 x축으로, 영어를 y축으로 하여 scatter를 그리고 각 원에 이름을 표시하시오
kor = df.kor
eng = df.eng
name = df.index
for h,w,n in zip(kor,eng,name):
    plt.text(h,w,'%s'%n)
plt.scatter(kor,eng,s=500,alpha=0.5) # 산점도 s는 사이즈임 리스트 줄수있어
plt.xlabel('국어')
plt.ylabel('영어')
plt.grid(True)
plt.show()
```
