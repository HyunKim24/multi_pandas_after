```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# 글꼴을 바꿔라
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
# 유니코드를 적용해라
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# data join 연습하기
# merge = PK랑 FK같이 연관된 컬럼이 있어야함

employee = [{'empno':1,'ename':'kim','dept':1},
           {'empno':2,'ename':'lee','dept':2},
           {'empno':3,'ename':'park','dept':1},
           {'empno':4,'ename':'song','dept':3},
           {'empno':5,'ename':'min','dept':2}
           ]
dept =[{'dept':1,'deptname':'관리직'},
       {'dept':2,'deptname':'영업직'},
       {'dept':3,'deptname':'개발직'}
      ]

info = [{'empno':1,'addr':'서울시','phone':'010-1111-1111'},
        {'empno':3,'addr':'부산시','phone':'010-2222-2222'},
        {'empno':2,'addr':'광주시','phone':'010-3333-3333'},
        {'empno':5,'addr':'광주시','phone':'010-4444-4444'},
        {'empno':4,'addr':'광주시','phone':'010-5555-5555'}
       ]
```

```python
dfemp = pd.DataFrame(employee)
dept = pd.DataFrame(dept)
info = pd.DataFrame(info)
```

```python
dfemp
```

```python
m = pd.merge(dfemp,dept, on='dept')
m.drop('dept',axis=1, inplace = True)
m
```

```python
pd.merge(m,info,on='empno')
```

```python
# concat - 결합할 데이터를 리스트로 줘야함
# row based concat vs column base concat
# axis =1 => 옆에다가 칼럼을 붙이면 됨
# axis =0 or default => 아래다가 붙임
pd.concat([dfemp,dept])
```

```python
pd.concat([dfemp,dfemp],axis=1)
```

```python
# join='outer' => 모든 로우를 다 살려서 함 =='DB OUTER JOIN'
pd.concat([dfemp,dept], axis=1, join='outer')
```

```python
#  join='inner' => NaN 값이 있는 로우는 삭제함 == 'DB INNER JOIN'
pd.concat([dfemp,dept], axis=1, join='inner')
```
