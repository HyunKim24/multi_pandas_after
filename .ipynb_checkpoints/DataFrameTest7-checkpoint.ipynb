```python
# read_table() => 텍스트 파일을 df로 변환하여 읽어오는 함수
# .read_table => 안에 있는 속성으로 적절히 데이터 변환하여 df에 쳐넣어
# 데이터 전처리 => df로 다 만들었음
df = pd.read_table('births.txt',sep=',',index_col =0,header = None)
df.columns = ['남아수','여아수']
df.index.name="년도"
df
```

#### <연습문제>
* 1. 남아수 총합을 구하시오
* 2. 여아수 평균을 구하시오
* 3. 남아수가 가장 많은 년도와 남아수를 구하시오
* 4. 2000년도 이후 데이터를 구하시오
* 5. 1995 ~ 2000년도 데이터를 구하시오
* 6. 남아 출생이 가장 많은 top5를 구하시오(년도와 남아수)
* 7. 남아율(%) 컬럼을 추가하고 남아수가 차지하는 비율을 보여주시오(예 50,70)
* 8. 2000년도 이후 데이터에 대해 출생량 컬럼을 추가하고 남아수가 1800000 이상이면 '많음' 아니면 '적음'이라고 보여주시오
* 9. 1800년대, 1900년대, 2000년대 각 남아,여아수 총합을 구하시오

```python
# 1. 남아수 총합을 구하시오
df['남아수'].sum()
```

```python
# 2. 여아수 평균을 구하시오
df['여아수'].mean()
```

```python
# 3. 남아수가 가장 많은 년도와 남아수를 구하시오
# df['남아수'].idxmax(),df['남아수'].max()
df[df['남아수']==df['남아수'].max()]['남아수']
```

```python
# 4. 2000년도 이후 데이터를 구하시오
df.loc[2000:]
```

```python
# 5. 1995 ~ 2000년도 데이터를 구하시오
df.loc[1995:2000]
```

```python
# 6. 남아 출생이 가장 많은 top5를 구하시오(년도와 남아수)
df.sort_values(by='남아수',ascending=False)[0:5]['남아수']
```

```python
# 7. 남아율(%) 컬럼을 추가하고 남아수가 차지하는 비율을 보여주시오(예 50,70)
df['남아율(%)'] = df['남아수']/df.sum(axis =1) * 100
df
```

```python
# 8. 2000년도 이후 데이터에 대해 출생량 컬럼을 추가하고 남아수가 1800000 이상이면 '많음' 아니면 '적음'이라고 보여주시
sr=df['남아수']
df['출생량']=sr.apply(lambda n :'많음' if n>1800000 else '적음')
df.loc['2000':,['남아수', '출생량']]
```

```python
# 9. 1800년대, 1900년대, 2000년대 각 남아수, 여아수 총합을 구하시오
df4 = [ df.loc['1800':'1899'].sum(), df.loc['1900':'1999'].sum(), df.loc['2000':].sum() ]
df4 = pd.DataFrame(df4, index=['1800년대', '1900년대', '2000년대'])
df4=df4[['남아수', '여아수']]
df4
```
