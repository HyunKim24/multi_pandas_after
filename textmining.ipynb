```python
url = 'https://rss.joins.com/joins_money_list.xml'
jfeed = feedparser.parse(url)
links = [entry['link'] for entry in jfeed['entries']] #link자리에 태그이름쓰면됨
links
```

```python
myTxt = ''
for lnk in links:
    article = newspaper.Article(lnk,language ='ko') # 접속하구
    article.download() #html 다운로드 받고
    article.parse() # 텍스트 가져오구
    myTxt+=article.text
    
myTxt[:500]
```

```python
len(myTxt)
```

```python
nData = t.nouns(myTxt)
```

```python
c = Counter(nData)
print(c)
nCount = c.most_common()
```

```python
fpath = 'C:\Windows\Fonts\malgun.ttf'
w = WordCloud(font_path=fpath, background_color='white',
             width=800,height=600)

cloud = w.generate_from_frequencies(c)
plt.figure(figsize=(12,8))
plt.imshow(cloud)
plt.show()
```
