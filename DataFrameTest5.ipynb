```python
import pandas as pd
import numpy as np
```

```python
data = {'kor':[10,30,50,70],'eng':[20,40,60,80],'math':[33,44,55,66]}
df = pd.DataFrame(data,index=['aa','bb','cc','dd'])
df
```

### 데이터 추가

```python
# sci라는 컬럼이 없기때문에 이걸 만들고 자료를 넣어줌
df['sci'] = [1,2,3,4]
df
```

```python
df['sci']=df['kor']+2
df
```

```python
# row의 개별데이터 끼리 더해서 새로운 컬럼을 만들어 집어 넣어라
df['sum'] = df['kor']+df['eng']+df['math']
df
```

```python
df
```

```python
# 해당 로우를 먼저 잡고 그 뒤 줄 값을 뒤에 리스트로 부여하면 됨
# 새로운 로우를 추가 할 경우 로우번호를 새롭게 주고 그 뒤 리스트로 새로운 값을 부여하면 됨
# df.loc['dd'] = [1,2,3]
df.loc['ee'] = [10,20,30]
df
```

```python
# df.iloc[0:,'avg'] = [1,2,3,4]
# df
df.loc['bb':,'avg'] = [2,3,4]
df
```

### 데이터 삭제

```python
# eng 컬럼이 삭제된 결과와 같다. 특정 컬럼을 선택하고 원래의 데이터에 주면
# 나머지 컬럼은 전부 삭제가 된다.
df = df[['kor','math']]
df
```

```python
df
```

```python
# 데이터 삭제는 데이터의 순서를 변경하는데에도 사용된다. 원하는 컬럼의 순서를 쓰고 그냥 원데이터에
# 받으면 원데이터의 컬럼 순서가 변경된다.
df = df[['math','eng','kor']]
df
```

```python
# 데이터 삭제는 drop 함수를 이용할 수 있음
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.drop.html
# drop(index)
# inplace=True 사용하면 원데이터를 안받아도됨
df.drop('bb',inplace=True)
df
```

```python
# 복수개를 삭제할 경우 리스트를 이용한다.
# df.drop(['aa','cc']) # df.loc['aa':'cc'] 이건 dd로우가 삭제된 결과와 같아여
# 컬럼이 드롭될 경우
# df.drop('kor',axis=1)
df.drop(['kor','eng'],axis=1)
```
