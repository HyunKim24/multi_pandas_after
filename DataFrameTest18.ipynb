```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# 글꼴을 바꿔라
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
# 유니코드를 적용해라
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
cctv = pd.read_csv('CCTV_in_Seoul.csv')
cctv
```

```python
#header=2,usecols='B,D,G,J,N') => 2번째 ROW부터 읽고 사용할 컬럼을 usecol에다가 적어라
pop = pd.read_excel('population_in_Seoul.xls',header=2,usecols='B,D,G,J,N')
pop.columns = ['기관명','인구총합','내국인총합','외국인총합','65세이상']
pop
```

```python
pop = pop.dropna()
pop
```

```python
cctv['증가율'] = (cctv['2014년']+cctv['2015년']+cctv['2016년'])/cctv['2013년도 이전']
cctv = cctv[['기관명','소계','증가율']]
cctv
```

```python
cctv.rename(columns={'소계':'cctv총수'},inplace=True)
cctv
```

```python
df = pd.merge(cctv,pop,on='기관명')
df
```

```python
df = df.set_index('기관명')
df
```

### 연습문제
* 1. cctv가 가장 많은 순으로 정렬하시오
* 2. 기관별 cctv 총수를 바차트(수평)로 표시 하시오
* 3. 기관별 cctv 증가율을 바차트(수평)로 표시하시오
* 4. 구별 외국인 비율과 고령자 비율 컬럼을 추가하시오
* 5. 기관별 인구대비 cctv 총수에 대한 cctv 비율 컬럼을 추가하시오
* 6. 기관별 cctv 비율을 바차트(수평)로 표시하시오

```python
# * 1. cctv가 가장 많은 순으로 정렬하시오
df1 = df.sort_values(by=['cctv총수'],ascending=False)
df1
```

```python
# * 2. 기관별 cctv 총수를 바차트(수평)로 표시 하시오
plt.barh(df.index, df['cctv총수'])
plt.show()
```

```python
# * 3. 기관별 cctv 증가율을 바차트(수평)로 표시하시오
plt.barh(df.index, df['증가율'])
plt.show()
```

```python
# * 4. 구별 외국인 비율과 고령자 비율 컬럼을 추가하시오
df4 = df
df4['외국인 비율 컬럼']=df4['외국인총합']/df4['인구총합']
df4['고령자 비율 컬럼']=df4['65세이상']/df4['인구총합']
df4
```

```python
# * 5. 기관별 인구대비 cctv 총수에 대한 cctv 비율 컬럼을 추가하시오
df5 = df
df5['인구대비 cctv 비율'] = df['cctv총수']  /df['인구총합']
df5
```

```python
#  6. 기관별 cctv 비율을 바차트(수평)로 표시하시오
plt.barh(df.index, df['인구대비 cctv 비율'])
plt.show()
```
