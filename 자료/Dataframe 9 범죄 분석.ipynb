```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
# 환경 설정
matplotlib.rcParams['font.family']='Malgun Gothic' # 제어판 글꼴 속성
matplotlib.rcParams['axes.unicode_minus'] = False # 유니코드를 적용해라
```

```python
crime = pd.read_csv('crime_in_Seoul.csv', engine='python', index_col='관서명')
crime.dropna( axis = 1)
```

```python
# 인코딩 문제 ->  engine='python': 파이썬의 인코딩 타입을 따르라는 의미
# index_col='관서명' 데이터의 일부분을 인덱스로 만들어 줌
crime = pd.read_csv('crime_in_Seoul.csv', engine='python', index_col='관서명', thousands=',') # ,를 제거
crime
```

```python
tmp = crime[ crime.columns[ crime.columns.str.contains('발생')]]
tmp['범죄발생수'] = tmp.sum(axis=1).values
tmp..dropna( axis =1)
```

```python
tmp2 = crime[ crime.columns[ crime.columns.str.contains('검거')]]
tmp2['검거수'] = tmp2.sum(axis=1).values
tmp2
```

```python
res = pd.DataFrame()
res['범죄발생수'] =tmp['범죄발생수']
res['검거수'] =tmp2['검거수']
res['검거율'] = tmp2['검거수']/tmp['범죄발생수']*100
res
```

```python
# 살인 범죄가 가장 심한 동네 top5
crime['살인 발생'].sort_values( ascending = False )[:5].plot(kind='bar')
plt.show()
```

```python
# 강도 범죄가 가장 심한 동네 top5
crime['강도 발생'].sort_values( ascending = False )[:5].plot(kind='bar', color='black')
plt.show()
```

```python
# 강간 범죄가 가장 심한 동네 top5
plt.xlim(250, 350)
crime['강간 발생'].sort_values( ascending = False )[:5].plot(kind='barh', color='red')
plt.show()
```

```python
# 절도 범죄가 가장 심한 동네 top5
plt.ylim(2000, 4000)
crime['절도 발생'].sort_values( ascending = False )[:5].plot(kind='bar', color='blue')
plt.show()
```

```python
# 폭력 범죄가 가장 심한 동네 top5
plt.ylim(2000, 4000)
crime['폭력 발생'].sort_values( ascending = False )[:5].plot(kind='bar', color='yellow')
plt.show()
```

```python
# 범죄 총 발생수 top5
plt.ylim(5000, 7000)
res['범죄발생수'].sort_values( ascending=False)[:5].plot(kind='bar')
plt.show()
```

```python
# 범죄 비율
data = [crime['살인 발생'].sum(), crime['강도 발생'].sum(), crime['강간 발생'].sum(), crime['절도 발생'].sum(), crime['폭력 발생'].sum()  ]
data2 = pd.Series( data )
index2 =(data2/data2.sum()*10000//1/100).values

# index1=['살인 %s%%' %index2[0] , '강도 %s%%'%index2[1], '강간 %s%%'%index2[2], '절도 %s%%'%index2[3], '폭력 %s%%'%index2[4]]
# index1
```

```python
index1 = ['살인', '강도', '강간', '절도', '폭력']
index3 = list()
len(index1)
for n in range( len( index1 ) ):
    index3.append(str(index1[n])+' '+str(index2[n])+'%')
```

```python
ratio = pd.Series(data, index=index3)
ratio.plot(kind='pie', title='범죄 비율', label='비율%')
plt.show()
```
