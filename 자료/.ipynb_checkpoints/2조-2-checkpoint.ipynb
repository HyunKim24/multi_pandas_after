```python
import pandas as pd
import numpy as n
import matplotlib.pyplot as plt
import matplotlib
import re
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']= False
```

```python
df=pd.read_csv('crime_in_Seoul.csv',index_col='관서명', thousands=',',  engine='python')


```

```python
df1 = df.iloc[0:,0::2]
df2=df.iloc[0:,1::2]
```

```python
# 지역별 검거율
df['검거율(%)'] =(df2.sum(axis=1)/df1.sum(axis=1))*100
df
```

```python
# 검거율 BEST3
df.sort_values(by='검거율(%)', ascending=False)[:3]
```

```python
# 지역별 범죄 발생률
df['지역별 범죄 발생률(%)']=(df1.sum(axis=1)/df1.sum(axis=1).sum())*100
df
```

```python
tmp=df['지역별 범죄 발생률(%)'].sort_values(ascending=False)[:10]
tmp.plot(kind='barh')
plt.show()
```

```python
# 지역별 범죄 발생률 TOP3
df.sort_values(by='지역별 범죄 발생률(%)', ascending=False)[:3]
```

```python
tmp=df['검거율(%)'].sort_values(ascending=False)[:10]
tmp.plot(kind='barh')
plt.show()
```

```python
new_df = pd.DataFrame(index=df.index )

cols = df.columns
for c1 in cols:
    for c2 in cols:
        if c1[0:2]==c2[0:2]:
            if ('발생' in c1) & ('검거' in c2):
                new_df[c1[0:2]] = (df[c2] / df[c1])*100
new_df
```

```python
new_df.plot(kind='barh')
plt.show()
```
