```python
from bs4 import BeautifulSoup
import urllib.request as REQ
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
urlbase = 'https://www.menupan.com'
urlsub = '/restaurant/bestrest/bestrest.asp?pt=rt&areacode=ss201'
response = REQ.urlopen(urlbase + urlsub)
soup = BeautifulSoup(response, 'html.parser')
```

```python
menuList=[]
hrefList=[]
ul = soup.select_one('.rankingList .list')
for li in ul.select('li'):
    href = li.select_one('.listName a')['href']       # listName 클래스 직계자손인 a의 href 속성
    hrefList.append(href)
    rank = li.select_one('.numTop, .rankNum').string  # 클래스가 numTop 이거나 rankNum  ,로 or구조 형성
    listName = li.select_one('.listName').string
    listType = li.select_one('.listType').string
    listArea = li.select_one('.listArea').string
    menuList.append({'랭킹':rank,
                     '상호명':listName, 
                     '업종':listType, 
                     '지역':listArea })
```

```python
menuList
```

```python
df = pd.DataFrame( menuList )
```

```python
hrefList
```

```python
telList=[]
addList=[]
urlbase = 'https://www.menupan.com'
for href in hrefList:   
    response = REQ.urlopen(urlbase + href)
    soup = BeautifulSoup(response, 'html.parser')
    telList.append( soup.select_one('.tel1').string )
    addList.append( soup.select_one('.add1').string )
    print( soup.select_one('.tel1').string)
    print( soup.select_one('.add1').string)
```

```python
df['전화번호'] = telList
df['주소'] = addList
```

```python
df = df.set_index('랭킹')
df
```

```python
df.to_excel('menu.xlsx')
```
