```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
df = pd.read_csv('crime_in_Seoul.csv',index_col='관서명',engine='python')
df
```

```python
import re
df[:].replace(regex=True,inplace=True,to_replace='\D',value='')
df
```

```python
# 1. 총 범죄가 가장 많이 발생한 지역
df=df.astype('int',inplace=True)
df['총범죄발생수']=df['살인 발생']+df['강도 발생']+df['강간 발생']+df['절도 발생']+df['폭력 발생']
df.sort_values(by ='총범죄발생수',ascending = False)[:5]['총범죄발생수']
```

```python
df[df.columns[0]][1]
```

```python
for r in range(5):
    for n in range(31):
        if df[df.columns[2*r]][n]<df[df.columns[2*r+1]][n]:
            df[df.columns[2*r+1]][n]=df[df.columns[2*r]][n]
df
```

```python
print('%s %s'%(df.columns[3] ,df.columns[2]))
```

```python
# 2. 범죄별 검거율
for n in range(5):
    print("%s / %s = "%(df.columns[2*n+1], df.columns[2*n]),df.sum()[2*n+1]/df.sum()[2*n]*100)
#     print(df.sum()[2*n+1]/df.sum()[2*n]*100)
```

```python
# 2-1 각 범죄 별 검거율
rate=list()
for n in range(5):
    tmp = df.sum()[2*n+1]/df.sum()[2*n]*100
    tmp = float(tmp)
    rate.append(tmp)
rate = pd.DataFrame(rate, index=['살인', '강도', '강간','절도','폭력'])
rate.columns=['범죄별 검거율']
rate
```

```python
# 2-2 차트
rate.plot(kind='bar')
plt.show()
```

```python
# 3. 검거율 높은 관서 top5
df['총범죄검거수']=df['살인 검거']+df['강도 검거']+df['강간 검거']+df['절도 검거']+df['폭력 검거']
df['총검거율']=df['총범죄검거수']/df['총범죄발생수']*100
df.sort_values(by ='총검거율',ascending = False)[:5]['총검거율']
```

```python
df2 = df.sort_values(by ='총검거율',ascending = False)[0:10]['총검거율']
df2.plot(kind='bar')
plt.show()
```

```python
# 4. 서별 강도 검거율
df['강도검거율']=df['강도 검거']/df['강도 발생']*100
```

```python
# 서별 범죄빈도수(많음 or 적음)
# df['총범죄발생수'].sum()/df.count()['살인 발생']
df['범죄빈도수']=df['총범죄발생수'].apply(lambda n: '많음' if n >=df['총범죄발생수'].sum()/df.count()['살인 발생'] else '적음')
df
```
