```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# thousands=',' => 천 단위로 콤마가 붙는 값을 숫자로 인식
df = pd.read_csv( 'crime_in_Seoul.csv', thousands=',', 
                 engine='python', index_col='관서명')
df[:5]
```

### 1. 관서별 범죄발생건수
* 관서별로 범죄(살인, 강도, 강간, 절도, 폭력) 발생 건수를 더한다
* 영등포서, 송파서, 관악서, 광진서, 마포서 순으로 범죄 발생 건수가 많았다

```python
df['범죄발생건수'] = df.iloc[0:, 0::2].sum( axis=1 )
df.sort_values(by='범죄발생건수', ascending=False)[:10]
```

```python
df['범죄발생건수'].sort_values(ascending=False)[:5].plot(kind='bar')
plt.show()
```

### 2. 관서별 중범죄 발생건수
* 관서별로 중범죄 발생건수를 더했다 
* 중범죄는 살인, 강간으로 정의했다
* 중범죄 발생건수는 서초서, 관악서, 영등포서, 강남서, 마포서 순으로 높았다

```python
df['중범죄 발생건수'] = df['살인 발생']+df['강간 발생']
df.sort_values('중범죄 발생건수', ascending=False)[:10]
```

```python
df['중범죄 발생건수'].sort_values(ascending=False)[:5].plot(kind='bar')
plt.show()
```

### 3. 관서별 검거 건수
* 관서별로 범죄 검거 건수를 더했다
* 영등포서, 송파서, 관악서, 광진서, 강서서 순으로 검거 건수가 많았다

```python
df['범죄검거건수'] = df.iloc[0:,1::2].sum( axis=1 )
df.sort_values('범죄검거건수', ascending=False)[:10]
```

```python
df['범죄검거건수'].sort_values(ascending=False)[:5].plot(kind='bar')
plt.show()
```

### 4. 관서별 범죄 검거율
* 관서별 검거율 = (서별검거건수 / 서별 범죄건수)*100
* 강서서, 금천서, 강북서, 도봉서, 수서서 순으로 범죄 검거율이 높았다

```python
df['서별 검거율'] = (df['범죄검거건수']/ df['범죄발생건수'])*100
df.sort_values(by='서별 검거율', ascending=False)[:10]
```

### 4-1. 서별 검거율 상위 5개
* 강서서(79.8%), 금천서(78.0%), 강북서(76.8%), 도봉서(75.2%), 수서서(73.6%)

```python
df['서별 검거율'].sort_values(ascending=False)[:5].plot(kind='bar')
plt.show()
```

### 4-2. 서별 검거율 하위 5개
* 중부서(63.7%), 관악서(63.6%), 강동서(63.5%), 방배서(63.1%), 남대문서(61.7%)

```python
df['서별 검거율'].sort_values(ascending=False)[-5:].plot(kind='bar')
plt.show()
```

### 5. 결론
* 관서별 위험지수 산출
* 위험지수 = (범죄발생건수 + (중범죄발생건수*0.3)) / 범죄검거건수

```python
df['위험지수'] = (df['범죄발생건수']+(df['중범죄 발생건수']*0.3)) / df['범죄검거건수']
df.sort_values(by='위험지수', ascending=False)[:10]
```

### 5-1. 위험지수 상위 5개
* 남대문서(1.63) > 방배서(1.60) > 관악서(1.59) > 강동서(1.58) > 중부서(1.58)

```python
df['위험지수'].sort_values(ascending=False)[:5].plot(kind='bar')
plt.show()
```

### 5-2. 위험지수 하위 5개
* 수서서(1.37) > 도봉서(1.34) > 강북서(1.31) > 금천서(1.29) > 강서서(1.27)

```python
df['위험지수'].sort_values(ascending=False)[-5:].plot(kind='bar')
plt.show()
```
