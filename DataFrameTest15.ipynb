```python
from bs4 import BeautifulSoup
import urllib.request as REQ 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
# 글꼴을 바꿔라
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
# 유니코드를 적용해라
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# url을 읽어와서 그거 연결시키는 부분
urlbase = 'http://www.menupan.com/'
urlsub = 'restaurant/bestrest/bestrest.asp?pt=rt&areacode=ss201'
response = REQ.urlopen(urlbase+urlsub)
# element 객체로 변환
soup = BeautifulSoup(response,'html.parser')
```

```python
# li.select_one('.numTop,.rankNum').string => selector에서 or를 표시할때는 이처럼 한다.
menuList = []
hrefList = []
ul = soup.select_one('.rankingList .list')
for li in ul.select('li'):
    
    # 속성 가지고 올때는 []안에 속성명을 준다.
    href = li.select_one('.listName a')['href']
    hrefList.append(href)
    
    rank = li.select_one('.numTop,.rankNum').string
    listName = li.select_one('.listName').string
    listType = li.select_one('.listType').string
    listArea = li.select_one('.listArea').string
    menuList.append({'랭킹':rank,'상호명':listName,'업종':listType,'지역':listArea})
```

```python
df = pd.DataFrame(menuList)
df
```

```python
telList = []
addList = []
# url을 읽어와서 그거 연결시키는 부분
urlbase = 'http://www.menupan.com/'
for href in hrefList:
    response = REQ.urlopen(urlbase+href)
    # element 객체로 변환
    soup = BeautifulSoup(response,'html.parser')
    telList.append(soup.select_one('.tel1').string)
    addList.append(soup.select_one('.add1').string)
```

```python
df['전화번호'] = telList
df['주소'] = addList
```

```python
df.to_excel("menu.xlsx")
```

```python
df = df.set_index('랭킹')
df
```
